## 00. パタトカクシー
2つの文字列「パトカー」と「タクシー」の文字を先頭から交互に連結し、文字列「パタトクカシーー」を得よ。

```python
s1     = "パトカー"
s2     = "タクシー"
result = "".join(c1 + c2 for c1, c2 in zip(s1, s2))
print(result)
```

## 03. 円周率
“Now I need a drink, alcoholic of course, after the heavy lectures involving quantum mechanics.”という文を単語に分解し、各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成せよ。

```python
import re
text = "Now I need a drink, alcoholic of course, after the heavy lectures involving quantum mechanics."
words = re.findall(r'[a-zA-Z]+', text)
word_lengths = [len(word) for word in words]
print(word_lengths)
```

## 04. 元素記号
“Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.”という文を単語に分解し、1, 5, 6, 7, 8, 9, 15, 16, 19番目の単語は先頭の1文字、それ以外の単語は先頭の2文字を取り出し、取り出した文字列から単語の位置（先頭から何番目の単語か）への連想配列（辞書型もしくはマップ型）を作成せよ。

```python
text  = "Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can."
words = re.findall(r'[a-zA-Z]+', text)

# 先頭1文字を取得する単語の位置（1-indexed）
single_char_positions = {1, 5, 6, 7, 8, 9, 15, 16, 19}

# 辞書を作成：抽出文字列 → 単語の位置
char_to_position = {}
for i, word in enumerate(words, 1):
    if i in single_char_positions:
        char = word[0]
    else:
        char = word[:2]
    char_to_position[char] = i

print(char_to_position)
```

## 05. n-gram
与えられたシーケンス（文字列やリストなど）からn-gramを作る関数を作成せよ。この関数を用い、”I am an NLPer”という文から文字tri-gram、単語bi-gramを得よ。

```python
def n_gram(sequence, n):

    if isinstance(sequence, str):
        # 文字列の場合→文字n-gram
        return [sequence[i:i+n] for i in range(len(sequence) - n + 1)]
    else:
        # リストやタプルなどのシーケンスの場合→単語n-gram
        return [sequence[i:i+n] for i in range(len(sequence) - n + 1)]# range()は終端を含まないので+1する

# 文
sentence = "I am an NLPer"

# 文字tri-gram
char_trigrams = n_gram(sentence, 3)
print("文字tri-gram:", char_trigrams)

# 単語bi-gram
words        = re.findall(r'[a-zA-Z]+', sentence)
word_bigrams = n_gram(words, 2)
print("単語bi-gram :", word_bigrams)
```

## 06. 集合
“paraparaparadise”と”paragraph”に含まれる文字bi-gramの集合を、それぞれXとYとして求め、XとYの和集合、積集合、差集合を求めよ。さらに、’se’というbi-gramがXおよびYに含まれるかどうかを調べよ。

```python
# 文字bi-gramの集合を作成
X = set(n_gram("paraparaparadise", 2))
Y = set(n_gram("paragraph", 2))
print("X (paraparaparadiseの文字bi-gram集合):", X)
print("Y (paragraphの文字bi-gram集合):", Y)
```

```python
# 和集合
union = X | Y
print("X ∪ Y (和集合):", union)
```

```python
# 積集合
intersection = X & Y
print("X ∩ Y (積集合):", intersection)
```

```python
# 差集合 (X - Y)
difference = X - Y
print("X - Y (差集合):", difference)
```

```python
# 'se' が含まれるか
print("'se' in X:", 'se' in X)
print("'se' in Y:", 'se' in Y)
```

## 08. 暗号文
与えられた文字列の各文字を、以下の仕様で変換する関数cipherを実装せよ。

・英小文字ならば (219 - 文字コード) のASCIIコードに対応する文字に置換  
・その他の文字はそのまま出力

この関数を用い、英語のメッセージを暗号化・復号化せよ。

```python
def cipher(text):
    result = []
    for char in text:
        if char.islower():
            # 英小文字の場合: (219 - 文字コード) のASCII文字
            result.append(chr(219 - ord(char)))
        else:
            # その他の文字はそのまま
            result.append(char)
    return ''.join(result)

# 英語のメッセージを暗号化
message   = "Hello World"
encrypted = cipher(message)
print("暗号化:", encrypted)

# 復号化（同じ関数で元に戻る）
decrypted = cipher(encrypted)
print("復号化:", decrypted)
```

## 09. Typoglycemia
スペースで区切られた単語列に対して、各単語の先頭と末尾の文字は残し、それ以外の文字の順序をランダムに並び替えるプログラムを作成せよ。ただし、長さが４以下の単語は並び替えないこととする。適当な英語の文（例えば”I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind .”）を与え、その実行結果を確認せよ。

```python
import random

def typoglycemia(text):
    words = text.split()
    result = []
    for word in words:
        if len(word) <= 4:
            result.append(word)
        else:
            first  = word[0]
            last   = word[-1]
            middle = [char for char in word[1:-1]]  # list() の代わりにリスト内包表記
            random.shuffle(middle)
            shuffled = first + ''.join(middle) + last
            result.append(shuffled)
    return ' '.join(result)

# 実行例
text   = "I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind ."
result = typoglycemia(text)
print(result)
```
