## 00. パタトカクシー
2つの文字列「パトカー」と「タクシー」の文字を先頭から交互に連結し、文字列「パタトクカシーー」を得よ。

```python
s1     = "パトカー"
s2     = "タクシー"
result = "".join(c1 + c2 for c1, c2 in zip(s1, s2))
print(result)
```

```python
s1     = "パトカー"
s2     = "タクシー"
```

```python
pairs  = list(zip(s1, s2))
print(pairs)
```

```python
list = [c1 + c2 for c1, c2 in pairs]
print(list)
```

```python
result = "".join(list)
print(result)
```

## 01. タクシー
文字列「パタトクカシーー」の2, 4, 6, 8文字目を取り出し、それらを連結した文字列を得よ。

```python
s      = "パタトクカシーー"
result = "".join(s[1::2])
print(result)
```
