## 13. タブをスペースに置換
ファイルの先頭10行に対して、タブ1文字につきスペース1文字に置換して出力せよ。確認にはsedコマンド、trコマンド、もしくはexpandコマンドなどを用いよ。

```python
with open("popular-names.txt", encoding="utf-8") as f:
    head_10 = f.readlines()[:10]

for line in head_10:
    print(line.replace("\t", " "), end="") # end="" は間の空行を消すため
```

### 13.1 sedコマンド  
sedコマンド:  
入力されたテキストデータを1行ずつ読み込んで指定した処理を適用して出力を行う。主に文字列の置換や抽出に用いられる。  

今回のような置換では以下のように使う。  
**text | sed 's/対象/置換語/g'**  

※オプションのgを用いることで、text内に含まれる**対象**全てに適用される。  

今回のコマンドは  
`head -10 popular-names.txt | sed 's/\t/ /g'`  

比較用  
`head -10 popular-names.txt | sed 's/\t/ /'`  


### 13.2 trコマンド
trコマンド: 文字を置換したり、削除したりするのに使う。 

今回のような置換では以下のように使う  
**text | tr '対象' '置換語'**  

今回のコマンドは    
`head -10 popular-names.txt | tr '\t' ' '`


### 13.3 expandコマンド

expandコマンド: テキスト内にあるタブを半角スペースに置き換えるコマンド  

今回のような置換では以下のように使う  
**text | expand -t 【数字】**  

※**-t 【数字】**で何個の半角スペースに置き換えるかを指定

今回のコマンドは  
`head -10 popular-names.txt | expand -t 1`
